```python
from pathlib import Path

from IPython.display import display

import atlas
from atlas.hrm.ops import init_hrm, load
from atlas.hrm.visualization import render_to_img

PROJECT_DIR = Path(atlas.__path__[0]).parent
```

```python
# Initialize an HRM with specific bounds
hrm = init_hrm(
    root_id=0,
    max_num_rms=20,
    max_num_states=5,
    max_num_edges=1,
    max_num_literals=5,
)

# Set the HRM path
hrm_file = "cw_4l_hrm_full.yaml"
path=PROJECT_DIR / "notebooks" / "hrms" / "data" / hrm_file

# Load the file into the HRM we have initialized
load(hrm, path)

# Render the HRM
img = render_to_img(hrm)
display(img)
```
